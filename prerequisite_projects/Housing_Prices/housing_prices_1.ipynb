# Importing Dataset And Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_url = "https://raw.githubusercontent.com/ibshafique/mlops_with_poridhi/refs/heads/main/prerequisite_projects/Housing_Prices/dataset/train.csv"
test_url = "https://raw.githubusercontent.com/ibshafique/mlops_with_poridhi/refs/heads/main/prerequisite_projects/Housing_Prices/dataset/test.csv"
```

```python
train_df = pd.read_csv(train_url)
test_df = pd.read_csv(test_url)
```

# Data Insights

Looking into the data statistically and visually for understanding more about the dataset.

```python
# Function to create scrollable table within a small window
def create_scrollable_table(df, table_id, title):
    html = f'<h3>{title}</h3>'
    html += f'<div id="{table_id}" style="height:200px; overflow:auto;">'
    html += df.to_html()
    html += '</div>'
    return html
```

# Cleaning Data

## Looking Into Value Counts

```python
for column in train_df.columns:
    print(f"Value counts for {column}:")
    print(train_df[column].value_counts())
    print()
```

## Looking Into Value Counts Graphically

```python
# Loop through each column and create a bar plot
for column in train_df.columns:
    # Get the value counts
    value_counts = train_df[column].value_counts()

    # Create a bar plot
    plt.figure(figsize=(8, 4))  # Adjust the figure size as needed
    value_counts.plot(kind='bar')

    # Add titles and labels
    plt.title(f"Value Counts for {column}")
    plt.xlabel("Values")
    plt.ylabel("Counts")

    # Show the plot
    plt.show()
```
